```python
import pandas as pd
import json
import os

# Initialize variables for chunk processing
file_path = '/home/code/library_api/authors/authors.json'
csv_output_path = 'authors.csv'
chunk_size = 10000  # Adjust this based on your memory limitations

def process_json_to_csv(json_file_path, csv_file_path, chunk_size):
    try:
        # Open JSON and CSV files
        with open(json_file_path, 'r') as json_file, open(csv_file_path, 'w') as csv_file:
            # Initialize variables
            header_written = False
            for i, line in enumerate(json_file):
                try:
                    # Parse JSON object
                    author_data = json.loads(line)
                    # Convert to DataFrame
                    author_df = pd.json_normalize(author_data)
                    
                    # Write DataFrame to CSV
                    author_df.to_csv(csv_file, index=False, header=not header_written, mode='a')
                    header_written = True

                    # Optionally, print progress every chunk
                    if (i + 1) % chunk_size == 0:
                        print(f"Processed {i + 1} lines")
                
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON on line {i + 1}: {e}")
                except Exception as e:
                    print(f"Error processing line {i + 1}: {e}")
                    
        print("CSV export complete!")

    except FileNotFoundError as e:
        print(f"File not found: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Run the function
process_json_to_csv(file_path, csv_output_path, chunk_size)
```
